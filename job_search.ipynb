### Minimum Salaries from £30k upwards

The results with the lowest minimum salaries all appear to be adverts for training courses not actual job roles.
I will filter these out by filtering the results to show results with a minimum salary greater than £30'000

```python
monthly_sals = df[df['minimumSalary']>30000]
monthly_sals.sort_values('minimumSalary', ascending=True).head(15)
```

### Number of job roles at each salary band

```python
monthly_sals = df[df['minimumSalary']>30000]
salary_bands = range(30000, int(df['minimumSalary'].max()) + 10000, 10000)

plt.figure(figsize=(10, 4))
plt.hist(monthly_sals['minimumSalary'], bins=salary_bands, edgecolor='black')
plt.xlabel('Minimum Salary (£)')
plt.ylabel('Number of Jobs')
plt.title('Number of jobs in each Salary Band')
plt.xticks(salary_bands)
plt.show()
```

```python
import nltk
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
def count_words(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text.lower())
    filtered_words = [word for word in words if word.isalpha() and word not in stop_words]
    return Counter(filtered_words)

# Apply the function to each job description
word_counts = df['jobDescription'].apply(count_words)

# Sum up the word counts across all job descriptions
total_word_counts = sum(word_counts, Counter())

# Create a DataFrame with the most common words
common_words_df = pd.DataFrame(total_word_counts.most_common(), columns=['Word', 'Count'])

# Display the table
print(common_words_df.head(50))
```
